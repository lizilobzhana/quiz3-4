1. (L6 – 3 ქულა) გამოიყენეთ მონაცემთა წმედის რომელიმე ფუნქციონალი (მაგ. რომელიმე სვეტის მონაცემები
გადაანაწილეთ რამდენიმე ახალ სვეტად, წაშალეთ რომელიმე სვეტი, ცარიელი მნიშვნელობებისთვის
გამოიყენეთ სასურველი ფორმატი, ა.შ.)

```python
df = df.drop_duplicates()
df
```

```python
df.drop(columns='Location Source', inplace=True)
df
```

```python
df[['Month', 'Day', 'year']] = df['Date'].str.split('/', expand=True)
df
```

```python
import numpy as np
df = df.replace(np.nan, '')
df
```

2. (L7 – 3 ქულა) რიცხვითი ტიპის ველის მონაცემები დაყავით კატეგორიებად cut() ფუნქციის გამოყენებით (მაგ.
ასაკის ველი 3 კატეგორიად „Children”, “Adolescent”, “Old”). შესაბამისად დაამატეთ კატეგორიის სვეტი.

```python
labels = ['Shallow', 'Moderate', 'Deep']
bins = [0, 50, 100, 700]  
df['Depth Category'] = pd.cut(df['Depth'], bins=bins, labels=labels, right=True)
df
```

3. (L9 – 3 ქულა) გამოიყენეთ მონაცემთა “Reshaping”-ის ჩამოთვლილი ფუნქციებიდან რომელიმე 2 მათგანი
(stack(), unstack(), melt(), pivot(), explode(), pivot_table(), Crosstab())

```python
df = df.drop_duplicates(['Date', 'Status'])
pivot_df = df.pivot(index='Date', columns='Status', values='Magnitude').reset_index()
pivot_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pt = df.pivot_table(index='year', columns=['Status', 'Depth Category'], aggfunc='mean', values='Depth')
pt
```

4. (L10 - 3 ქულა) ააგეთ თქვენთვის სასურველი 4 განსხვავებული დიაგრამა შემდეგი ტიპებიდან (ხაზოვანი
დიაგრამა, სვეტოვანი დიაგრამა, გაფანტულობის დიაგრამა, წრიული დიაგრამა, ჰისტოგრამა).

```python
fig =  plt.figure(figsize=(15,5))
sns.lineplot(x='Date', y='Magnitude', data=df, marker='o', linestyle='--')
plt.title('Earthquake Magnitudes Over Time')
plt.xlabel('Date')
plt.ylabel('Magnitude')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Status', data=df, palette='Set2')
plt.title('Earthquake Counts by Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Magnitude'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
gender_counts = df['Status'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Status Distribution')
plt.show()
```

5. (L11 – 2 ქულა) გამოიყენეთ თარიღებთან სამუშაო მინიმუმ 2 განსხვავებული ფუნქცია/ტრანსფორმაცია
სურვილისამებრ (მაგ. resample ფუნქცია).

```python
data = {
    'Date': ['1/2/1965', '1/4/1965', '1/5/1965', '1/8/1965', '1/9/1965'],
    'Magnitude': [6.0, 5.8, 6.2, 5.8, 5.8],
    'Status': ['Automatic', 'Automatic', 'Automatic', 'Reviewed', 'Reviewed'],
    'Latitude': [19.2460, 1.8630, -20.5790, -59.0760, 11.9380],
    'Longitude': [145.6160, 127.3520, -173.9720, -23.5570, 126.4270],
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df.set_index('Date', inplace=True)
```

```python
magnitude_weekly = df.resample('W')['Magnitude'].mean()
magnitude_weekly
```

```python
df.resample('W').agg({'Magnitude': 'sum', 'Latitude': 'max', 'Longitude': 'min'})
```
